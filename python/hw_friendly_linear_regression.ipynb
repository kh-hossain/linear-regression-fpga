```python
#!/usr/bin/env python
# coding: utf-8

# In[1]:

# Enter Y and X (hardware implementation ideas)
'''
x = [] # size 200 * 10 = 2000
y = [] # size 200 * 1  = 200

i = 0
j = 0
k = 0
while (i != 200):
    y[i] = float(input())
    while (j != 10)
        x[k] = float(input())
        j++
        k++
    i++
    j = 0
'''

#################################################################

import random

import numpy as np

# Generate random coefficients
numOfBetas = 0
beta = [0] * 11

random.seed(1)

while (numOfBetas != 11):
    beta[numOfBetas] = (random.random() + 0.5) * (random.random() * 5)
    numOfBetas = numOfBetas + 1

# Generate x values 
x = [0] * 2000
i = 0
j = 0
count = 0

while (i != 200):
    j = 0
    
    while (j != 10):
        x[count] = (random.random() + 0.5) * (random.random() * 10)
        count = count + 1
        j = j + 1
        
    i = i + 1


# Generate y values
y = [0] * 200
i = 0
j = 0
count = 0

while (i != 200):
    y_i = beta[0]
    j = 0
    
    while (j != 10):
        y_i = y_i + beta[j+1] * x[count]
        count = count + 1
        j = j + 1
        
    y[i] = y_i
    i = i + 1

#################################################################

# Compute sum(y) and store in first row of XtY

xty = [0] * 11
addrY = 0
result = 0

while (addrY != 200):
    result = result + y[addrY]
    addrY += 1
    
xty[0] = result

# Compute the rest of XtY

addrXtY = 1 # Skip first row, already calculated
addrY = 0
addrX1 = 0

while (addrXtY != 11):
    result = 0
    
    while (addrY != 200): # Calculate sum
        result = result + y[addrY] * x[addrX1]
        addrY += 1
        addrX1 += 10
        
    xty[addrXtY] = result
    addrX1 = addrXtY
    addrXtY += 1
    addrY = 0

#################################################################
    
xtx = [0] * 121
xtx[0] = 200 # Number of n (num of samples)

# Compute first row and coloumn of XtX

horizontal = 1
vertical = 11
addrX1 = 0

while (horizontal != 11):
    result = 0
    count = 0
    
    while (count != 200): # Compute sum
        result = result + x[addrX1]
        addrX1 += 10
        count += 1
        
    xtx[horizontal] = result
    xtx[vertical] = result
    addrX1 = horizontal
    horizontal += 1
    vertical += 11

    
# Compute the rest of XtX.

indexXtX = 12 # start at location of XtX[(x_i1 ^ 2)] i.e. second row second column
indexRow = 0 # for the second row of XtX, start at xi1. Eventually, indexRow=9 i.e. xi10 for the last row.
indexColumn = 0 # for each row of XtX, when indexColumn=10 stop. We reached the final column. Go to a new row.

while (indexXtX != 132): # Row loop # 121 + 12 = 132, means that we reached out side of the index and already computed element 121
    
    horizontal = indexXtX
    vertical = indexXtX
    indexColumn = indexRow
    while (indexColumn != 10): # Column loop
        
        addrX1 = indexRow
        addrX2 = indexColumn
        result = 0
        count = 0
        while (count != 200): # Calculate sum
            result = result + x[addrX1] * x[addrX2]
            addrX1 += 10
            addrX2 += 10
            count += 1

            
        xtx[horizontal] = result
        xtx[vertical] = result
        horizontal += 1
        vertical += 11
        indexColumn += 1
        
    # move to next row
    indexXtX += 12
    indexRow += 1  
    
print("*" * 50)
print("xtx[45]: ",xtx[45])
print("xtx[23]: ",xtx[23])
print("xtx[120]: ",xtx[120])

print("*" * 50)
print("xty[10]: ",xty[10])
print("xty[5]: ",xty[5])
print("xty[2]: ",xty[2])
#################################################################

# Matrix inversion

# Create an identity matrix
inverse = [0] * 121
i = 0

while (i != 132):
    inverse[i] = 1
    i += 12

# Perform Gauss-Jordan elimination
i = a = b = c = e = f = 0

while i != 11: # Divide the current row by the pivot element
    
    pivot = xtx[a]
    counter = 0
    
    while counter != 11:
        xtx[b] /= pivot
        inverse[b] /= pivot
        counter += 1
        b += 1

    # Subtract multiples of the pivot row from other rows
    d = 0
    c = i
    counter1 = 0
    
    while counter1 != 11:
        
        if i != counter1:
            factor = xtx[c]
            e = f
            counter2 = 0
            
            while counter2 != 11:
                xtx[d] -= factor * xtx[e]
                inverse[d] -= factor * inverse[e]
                e += 1
                d += 1
                counter2 += 1
                
        else: d += 11
            
        counter1 += 1
        c += 11

    f += 11
    i += 1
    a += 12


# Final matrices multiplication: (XtX) ^ -1 * XtY

a = [0] * 11

# Initialize index variables
addrInv = 0
addrXtY = 0
addrA = 0

# Multiply the elements of xtx and xty and store the result in (a)
while addrInv != 121: # Final item has index 121
    a[addrA] = a[addrA] + inverse[addrInv] * xty[addrXtY]

    # Increment the index variables
    addrInv = addrInv + 1
    addrXtY = addrXtY + 1

    # Move to beggining of xty if j reaches 11 (i.e. the end of it)
    if addrXtY == 11:
        addrXtY = 0
        addrA = addrA + 1



# print("Original betas (coeffiecients used for data generation):\n", beta)
# print("*" * 50)
# print("Program generated betas (final matrix A):\n", a)
```
