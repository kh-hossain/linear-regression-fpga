```python

#################################################################

import random

import numpy as np

# Generate random coefficients

random.seed(1)

numOfBetas = 0
beta = [0] * 11

while (numOfBetas != 11):
    beta[numOfBetas] = (random.random() + 0.5) * (random.random() * 5)
    numOfBetas = numOfBetas + 1

# Generate x values 
x = [0] * 2000
i = 0
j = 0
count = 0

while (i != 200):
    j = 0
    
    while (j != 10):
        x[count] = (random.random() + 0.5) * (random.random() * 10)
        count = count + 1
        j = j + 1
        
    i = i + 1


# Generate y values
y = [0] * 200
i = 0
j = 0
count = 0

while (i != 200):
    y_i = beta[0]
    j = 0
    
    while (j != 10):
        y_i = y_i + beta[j+1] * x[count]
        count = count + 1
        j = j + 1
        
    y[i] = y_i
    i = i + 1

#################################################################

import struct

def convert_to_binary(input_data, output_file_path):
    with open(output_file_path, 'wb') as output_file:
        for element in input_data:
            # Convert the string representation to a floating-point number
            float_number = float(element)
            
            # Convert the float to binary representation and write to the output file
            binary_representation = struct.pack('!f', float_number)
            output_file.write(binary_representation)

# Example list of string representations of floating-point numbers
input_data_list_x = x.copy()
input_data_list_y = y.copy()

# Specify the path for the output file
output_file_path_x = 'output_binary_data_x_STRUCT.txt'
output_file_path_y = 'output_binary_data_y_STRUCT.txt'

# Perform the conversion
convert_to_binary(input_data_list_x, output_file_path_x)
convert_to_binary(input_data_list_y, output_file_path_y)


print(f"Conversion complete. Binary data saved to '{output_file_path_x}'.")
print(f"Conversion complete. Binary data saved to '{output_file_path_y}'.")
```
